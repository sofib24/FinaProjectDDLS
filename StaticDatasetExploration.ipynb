```python
import os
import matplotlib.pyplot as plt
from PIL import Image

mip_folder = 'data/MIP'
segmentation_folder = 'data/Segmentations'
# Hardcoded values for condition and timepoint
condition = 'A'
timepoint = '24h'
image_number = '005' # Example image number

# Construct the image name from the hardcoded values
image_name = f'{condition}_{timepoint}_Image{image_number}_merge_all.tif'
image_path = os.path.join(mip_folder, image_name)

# Construct the mask name from the hardcoded values
mask_name = f'{condition}_{timepoint}_Image{image_number}.tiff'
mask_path = os.path.join(segmentation_folder, mask_name)

# Check if the image and mask files exist
if os.path.exists(image_path) and os.path.exists(mask_path):
    # Open the image and mask
    image = Image.open(image_path)
    mask = Image.open(mask_path)
    
    # Create a figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
    
    # Display the image on the first subplot
    ax1.imshow(image)
    ax1.set_title(f'Image: {condition} {timepoint}')
    ax1.axis('off')
    
    # Display the mask on the second subplot
    ax2.imshow(mask, cmap='gray')
    ax2.set_title(f'Mask: {condition} {timepoint}')
    ax2.axis('off')
    
    # Save the figure
    plt.savefig('plot.png')
else:
    # Print an error message if the image or mask is not found
    if not os.path.exists(image_path):
        print(f'Image not found: {image_path}')
    if not os.path.exists(mask_path):
        print(f'Mask not found: {mask_path}')
```
